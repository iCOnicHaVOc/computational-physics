```python
#Question 3
from My2lib import *
P_x = LCG(5600)
P_y = LCG(1000)

def check(x_list,y_list):
    Pi_n =[]
    index=[]
    total_points = len(x_list)
    inside_pi=0
    for idx,(x,y) in enumerate(zip(x_list,y_list), start=1):
        if x**2+y**2<=1:
            inside_pi +=1
        pi_est = 4*(inside_pi/idx)
        Pi_n.append(pi_est)
        index.append(idx)
    return index,Pi_n

p,q = check(P_x,P_y)
Plot(p,q)
```
